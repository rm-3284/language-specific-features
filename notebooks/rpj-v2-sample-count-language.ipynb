```python
import re
from collections import defaultdict
from tqdm.auto import tqdm
```

```python
from datasets import load_dataset

ds = load_dataset("EleutherAI/rpj-v2-sample", streaming=True)
```

```python
ds["train"]
```

```python
lang_counts = defaultdict(int)

for i, example in tqdm(enumerate(ds["train"])):
    match = re.search(r'(?<=\/).{2}(?=_)', example["doc_id"])
    
    if match:
        lang = match.group()
        lang_counts[lang] += 1
```

```python
lang_counts
```

```python
for lang, count in lang_counts.items():
    print(f"{lang}: {count / sum(lang_counts.values()) * 100:.2f}%")
```
